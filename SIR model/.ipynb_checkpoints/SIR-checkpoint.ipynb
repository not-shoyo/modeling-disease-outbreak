```python
import pandas as pd
import numpy as np
import random
import copy
pd.options.mode.chained_assignment = None

meta = pd.read_csv('metadata.csv', sep = '\t', names=["ID", "Class", "Gender"])
primary = pd.read_csv('primaryschool.csv', sep = '\t', names=["Time", "ID1", "ID2", "Class1", "Class2"])
# contact_network = pd.read_csv('primaryschool.csv', sep='\t', lineterminator='\n', names=['Time', 'ID 1', 'ID 2', 'Class 1', 'Class 2']).sort_values(by=['Time'])


primary
```

```python
N = len(primary["Time"].unique())
print("Unique Time Stamps:", N)
N = len(meta["ID"].unique())
print("Unique IDs in meta", N)
```

```python
#3100 unique time stamps => assume 100 time stamps make 1 day => 31 days (to be able to simulate over 1 month)
```

```python
class SIR:
    initial_infected = int (0.06 * N)
    beta = 0.005
    gamma = 0.0002
    
    def __init__(self, primary, meta):
        self.meta = meta
        self.primary = primary
    
    def simulate(self):
        self.meta["State"] = "S" 
#         self.primary["Time"] = self.primary["Time"] - self.primary["Time"][0]
#         self.primary["Time"] = self.primary["Time"] / 20
        initial_indexes = random.sample(list(self.meta["ID"]), self.initial_infected)
        for i in initial_indexes:
            self.meta["State"][self.meta.loc[self.meta['ID'] == i].index[0]] = "I"
        Y = [self.initial_infected]
        X = [len(self.meta) - self.initial_infected]
        Z = [0]
        print(X, Y, Z)
        T = [self.primary["Time"][0]]
#         T = []
        i, stamps, pLen = 0, 0, len(primary)
        infections, recoveries = 0, 0
        while i < pLen:
#             print(i, stamps)
            if stamps >= 100:
                recoveries = 0
                for idx in self.meta.index:
                    if self.meta["State"][idx] == "I" and random.random() < self.gamma:
                        recoveries += 1
                        self.meta["State"][idx] = "R"
                print(infections, recoveries)
#                 print(X[-1] - infections, Y[-1] + infections - recoveries, Z[-1] + recoveries)
                X.append(X[-1] - infections)
                Y.append(Y[-1] + infections - recoveries)
                Z.append(Z[-1] + recoveries)
                infections = 0
                temp = (T[-1] - 31220) // 20 // 100
                print(f"Day {temp} overr")
#                 print(f"X: {X[-1]}, Y: {Y[-1]}, Z: {Z[-1]}")
                stamps = 0
            else:
                stamps += 1  
            T.append(self.primary["Time"][i])
            while i < pLen and self.primary["Time"][i] == T[-1]:
                id1, id2 = self.primary["ID1"][i], self.primary["ID2"][i]
                if self.meta["State"][self.meta.loc[self.meta['ID'] == id1].index[0]] == "I" and self.meta["State"][self.meta.loc[self.meta['ID'] == id2].index[0]] == "S":
                    if random.random() < self.beta:
                        infections += 1
                        self.meta["State"][id2] = "I"
                elif self.meta["State"][self.meta.loc[self.meta['ID'] == id1].index[0]] == "S" and self.meta["State"][self.meta.loc[self.meta['ID'] == id2].index[0]] == "I":
                    if random.random() < self.beta:
                        infections += 1
                        self.meta["State"][id1] = "I"
                i += 1
        return X, Y, Z, T

model = SIR(meta=meta, primary=primary)
    
```

```python
X, Y, Z, time = model.simulate()
```

```python
S = np.array(X) / N
I = np.array(Y) / N
R = np.array(Z) / N
T = np.array(range(31))

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.plot(T, S, label='Susceptible')
plt.plot(T, I, label='Infected')
plt.plot(T, R, label='Recovered')
plt.legend()
plt.title('Disease Spread')
plt.xlabel('Time')
plt.ylabel('Fraction of Population')
plt.show()
```

```python
class SIR_model:
    
    def __init__(self, beta, gamma, start_infected):
        self.beta=beta
        self.gamma=gamma
        self.start_infected=start_infected
    
    def introduce_infection(self, start_infected):
        res = random.sample(range(0, N), start_infected)
        for id in res:
            status[id] = 'I'
#         for i in range(start_infected):
#             infected=random.choice(student_data['ID'])
#             compartment[infected-start_id] = 'I'

    def recovery(self,R,I):
        recoveries = R
        infections = I
        for id in meta['ID']:
            if status[list(meta['ID']).index(id)] == 'I':
                if random.random() < self.gamma:
                    status[list(meta['ID']).index(id)] = 'R'
                    recoveries += 1
                    infections -= 1
        return recoveries, infections
#         recoveries=R
#         infected=I
#         for id in student_data['ID']:
#             if compartment[id-start_id] == 'I':
#                 if random.random()< self.gamma:
#                     compartment[id-start_id] == 'R'
#                     recoveries+=1
#                     infected-=1
#         return recoveries,infected

    def simulate(self):
        self.introduce_infection(self.start_infected)
        R=0
        I=self.start_infected
        S=N-I-R
        X = [S]
        Y = [I]
        Z = [R]
        Time = [0]
        ActualX, ActualY, ActualZ = [], [], []
        ActualTime = []
        i, count = 0, 0
        print(primary.shape[0])
        while i < primary.shape[0]:
            t = Time[-1] + 10
            while i < primary.shape[0] and primary['Time'][i] == t:
                id1 = primary['ID1'][i]
                id2 = primary['ID2'][i]
                if status[list(meta['ID']).index(id1)] == 'S' and status[list(meta['ID']).index(id2)] == 'I':
                    if random.random() < self.beta:
                        S -= 1
                        I += 1
                        status[list(meta['ID']).index(id1)] = 'I'
                elif status[list(meta['ID']).index(id2)] == 'S' and status[list(meta['ID']).index(id1)] == 'I':
                    if random.random() < self.beta:
                        S -= 1
                        I += 1
                        status[list(meta['ID']).index(id2)] = 'I'
                i = i + 1
            if count == 0 or i == primary.shape[0] or (Time[-1] in list(primary['Time']) and i != 0 and i != primary.shape[0] and primary['Time'][i] != primary['Time'][i-1]):
                count += 1
                if count % 100 == 0:
                    ActualX.append(S)
                    ActualY.append(I)
                    ActualZ.append(R)
                    ActualTime.append(count // 100)
#             elif count == 0 or i == primary.shape[0]:
#                 count += 1
            R,I = self.recovery(R,I)

            X.append(S)
            Y.append(I)
            Z.append(R)
            Time.append(t)

        print(count)
        print(ActualX, len(ActualY), ActualZ, ActualTime)
        return ActualX, ActualY, ActualZ, ActualTime        
#         return X, Y, Z, Time
```

```python
m = SIR_model(
    beta = 1/200, #30 #200
    gamma = 1/2500, #5000
    start_infected = 15 #50 #15
)

X, Y, Z, t = m.simulate()
```

```python
Rate_S = np.array(X)/N
Rate_I = np.array(Y)/N
Rate_R = np.array(Z)/N
time = np.array(t)
# time = np.array(t) / (60 * 24)

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
plt.plot(time, Rate_S, label='Susceptible')
plt.plot(time, Rate_I, label='Infected')
plt.plot(time, Rate_R, label='Recovered')
plt.xticks(range(32))
plt.legend()
plt.title('Disease Spread (SIR Model)')
plt.xlabel('Timeline')
plt.ylabel('Fraction of Population')
plt.show()
```
